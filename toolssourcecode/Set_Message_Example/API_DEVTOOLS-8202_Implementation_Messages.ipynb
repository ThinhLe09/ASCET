```python
from win32com import client
ascet = client.Dispatch('Ascet.Ascet.6.1.4')
```

```python
db = ascet.GetCurrentDataBase() #db: DataBaseItem
Folders_in_DB = db.GetAllFolders()
Folder_Root = db.GetFolder(Folders_in_DB[0].GetName())
```

```python
CC = db.GetItem('CC')
```

```python
bb = CC.GetModelElement('bb')
cc = CC.GetModelElement('cc')
dd = CC.GetModelElement('dd')
```

```python
#ImplItem_aa = aa.GetImplementation()
ImplItem_bb = bb.GetImplementation()
ImplItem_cc = cc.GetImplementation()
ImplItem_dd = dd.GetImplementation()


#ImplInfo_aa = ImplItem_aa.GetImplInfoForValue()
ImplInfo_bb = ImplItem_bb.GetImplInfoForValue()
ImplInfo_cc = ImplItem_cc.GetImplInfoForValue()
ImplInfo_dd = ImplItem_dd.GetImplInfoForValue()


print('bb formulae:', ImplInfo_bb.GetFormulaName())
print('bb impl type:', ImplInfo_bb.GetImplType())
print('bb impl range:', ImplInfo_bb.GetDoubleImplRange())
print('bb phy range:', ImplInfo_bb.GetDoublePhysicalRange())
print()
print('cc formulae:', ImplInfo_cc.GetFormulaName())
print('cc impl type:', ImplInfo_cc.GetImplType())
print('cc impl range:', ImplInfo_cc.GetDoubleImplRange())
print('cc phy range:', ImplInfo_cc.GetDoublePhysicalRange())
print()
print('dd formulae:', ImplInfo_dd.GetFormulaName())
print('dd impl type:', ImplInfo_dd.GetImplType())
print('dd impl range:', ImplInfo_dd.GetDoubleImplRange())
print('dd phy range:', ImplInfo_dd.GetDoublePhysicalRange())
```
